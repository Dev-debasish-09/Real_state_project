```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('corrected_data.csv')
```

```python
df.head()
```

```python
import seaborn as sns
plt.figure(figsize = (4,3))
sns.boxplot(df['luxury_score'],orient = 'h')
```

```python
def luxury_catagory(score):
    if 0<= score <50:
        return 'Low'
    elif 50<= score <150:
        return 'Medium'
    elif 150<=score<=175:
        
        return 'High'
    else:
        return None
```

```python
df['luxury_category'] = df['luxury_score'].apply(luxury_catagory)
```

```python
df.head()
```

```python
new_df = df.drop(columns=['luxury_score'],axis = 1)
```

```python
new_df.head()
```

```python
plt.figure(figsize=(5,3))
sns.boxplot(df['floorNum'],orient = 'h')
```

```python
def floor_cat(value):
    if 0<= value <=2:
        return "Low Floor"
    elif 3<= value <= 10:
        return "Med Floor"
    elif 11 <= value<=51:
        return "High Floor"
    else:
        return None
```

```python
new_df['Floor_category'] = new_df['floorNum'].apply(floor_cat)
```

```python
new_df.head()
```

```python
new_df = new_df.drop(columns=['floorNum'],axis = 1)
```

```python
new_df.head()
```

```python
#label Encoder
from sklearn.preprocessing import OrdinalEncoder
data_label= new_df.copy()
categorical_cols = new_df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    OEncoder = OrdinalEncoder()
    data_label[col] = OEncoder.fit_transform(data_label[[col]])
```

```python
categorical_cols = new_df.select_dtypes(include=['object']).columns
```

```python
categorical_cols
```

```python
data_label.head()
```

```python
x = data_label.drop('price',axis=1)
y = data_label['price']
```

```python
x
```

```python
y
```

```python
# feature selection
# corelation analysis
plt.figure(figsize=(6,4))
sns.heatmap(data_label.corr())
```

```python
co = data_label.corr()['price'].iloc[1:].to_frame().reset_index().rename(columns={'index':'feature','price':'corr_coef'})
co
```

```python
# random forest feature selection
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators =100,random_state=42)
rf.fit(x,y)
co_2 = pd.DataFrame({
    'feature': x.columns,
    'feature_importance':rf.feature_importances_
}).sort_values(by='feature_importance',ascending=False)
co_2
```

```python
#gradient boosting feature importance
from sklearn.ensemble import GradientBoostingRegressor

gb = GradientBoostingRegressor()
gb.fit(x,y)
co_3 = pd.DataFrame({
    'feature': x.columns,
    'gb_feature_importance':gb.feature_importances_
}).sort_values(by = 'gb_feature_importance',ascending=False)
co_3
```

```python
# permutation importance
from sklearn.inspection import permutation_importance
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state =42)
rf_1 = RandomForestRegressor(n_estimators = 100,random_state = 42)
rf_1.fit(x_train,y_train)

perm_importance = permutation_importance(rf_1,x_test,y_test,n_repeats = 30,random_state = 42)

co_4 = pd.DataFrame({
    'feature':x.columns,
    'permutation_importance':perm_importance.importances_mean
}).sort_values(by='permutation_importance',ascending=False)

co_4
```

```python
#lasso 
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaler = scaler.fit_transform(x)
lasso = Lasso(alpha=0.01,random_state=42)
lasso.fit(x_scaler,y)
co_5 = pd.DataFrame({
    'feature':x.columns,
    'lasso_coef':lasso.coef_
}).sort_values(by='lasso_coef',ascending=False)

co_5
```

```python
#rfe (recursive feature elimination)
from sklearn.feature_selection import RFE
estimator = RandomForestRegressor()

selector = RFE(estimator,n_features_to_select=x.shape[1],step=1)
selected = selector.fit(x,y)

selected_features = x.columns[selector.support_]
selected_coefficients = selector.estimator_.feature_importances_

co_6 = pd.DataFrame({
    'feature':selected_features,
    'RFE_score':selected_coefficients
}).sort_values(by='RFE_score',ascending=False)

co_6
```

```python
f_df = co.merge(co_2,on='feature').merge(co_3,on='feature').merge(co_4,on='feature').merge(co_5,on='feature').merge(co_6,on='feature').set_index('feature')
```

```python
f_df
```

```python
final_norm_df = f_df.divide(f_df.sum(axis=0),axis=1)
```

```python
final_norm_df
```
